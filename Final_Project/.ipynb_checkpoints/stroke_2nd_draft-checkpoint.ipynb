```python
# drop the one person with other gender
stroke_df.drop(stroke_df.index[3116], axis=0, inplace=True)
```

```python
over_25_df = pd.DataFrame(stroke_df[stroke_df['age'] >= 25])
```

```python
over_25_df.info()
```

```python
just_stroke = over_25_df[over_25_df["stroke"] == 1]
just_stroke
```

```python
just_stroke.describe()
```

```python
# Create bins and bin labels for the Age column

age_bins = [25, 35, 45, 55, 65, 75, 85]
age_labels = ["25s", "35s", "45s", "55s", "65s", "75s"]

# Bin the age column
# cut() returns a Pandas Series containing each of the binned column's values translated into their corresponding bins

pd.cut(over_25_df['age'], age_bins, labels=age_labels)
pd.cut(just_stroke["age"], age_bins, labels=age_labels)
```

```python
over_25_df['age_groups'] = pd.cut(over_25_df['age'], age_bins, labels=age_labels)
just_stroke['age_groups'] = pd.cut(just_stroke["age"], age_bins, labels=age_labels)
just_stroke.head()
```

```python
# subset people who had strokes who aren't missing BMI scores
stroke_no_null = just_stroke.dropna(axis =0)
stroke_no_null
```

```python
# Find averages of BMIs by age group for those who had strokes
twentys_bmi= stroke_no_null[stroke_no_null['age_groups'] == "25s"]['bmi'].mean() 
twentys_bmi
```

```python
thirtys_bmi= stroke_no_null[stroke_no_null['age_groups'] == "35s"]['bmi'].mean() 
thirtys_bmi
```
