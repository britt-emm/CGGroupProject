```python
import pandas as pd
import numpy as np

fire_df = pd.read_csv('fire_data.csv')
fire_df.head()
```

```python
fire_df.columns
```

```python
fire_df['Hemi_ID'] = fire_df['Unnamed: 0']
fire_df.head()
```

```python
fire_df.drop('Unnamed: 0', axis=1, inplace=True)
fire_df.head()
```

```python
fire_df.info()
```

```python
fire_df['End Date'] = pd.to_datetime(fire_df['End Date'])
fire_df['Event Date'] = pd.to_datetime(fire_df['Event Date'])
fire_df.info()
```

```python
fire_df['Event Length'] = fire_df['End Date'] - fire_df['Event Date']
fire_df.head()
```

```python
fire_subset = pd.DataFrame(fire_df[['ID','Name of Fire', 'End Date', 'Event Date', 'Year', 'Month', 'Event Length']])
country_subset = pd.DataFrame(fire_df[['ID', 'Latitude','Longitude', 'Country', 'Hemi_ID']])
hemisphere_subset = pd.DataFrame(fire_df[['Hemi_ID', 'Hemisphere']])
```
